```python
# how to check which states have the highest churn rate
churn_rate_by_state = df.groupby('State')['Churn'].mean().sort_values(ascending=False)
print(churn_rate_by_state)
```

```python
# how to check which states have the highest minutes called
avg_minutes_by_state = df.groupby('State')['Total day minutes'].mean().sort_values(ascending=False)
print(avg_minutes_by_state)
```

```python
df['Total minutes'] = (
    df['Total day minutes'] +
    df['Total eve minutes'] +
    df['Total night minutes'] +
    df['Total intl minutes']
)

df.groupby('State')['Total minutes'].sum().sort_values(ascending=False)
```
